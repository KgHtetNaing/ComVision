```python
import cv2 as cv
import matplotlib.pyplot as plt
print(cv.__version__)
```

```python
img = cv.imread("example_02.png")
img2 = cv.imread ("example_01.png")
```

```python
img.shape
```

```python
plt.imshow(img2)
```

```python
gray = cv.cvtColor(img2 , cv.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
```

```python
blur_img = cv.GaussianBlur(gray , (11,11) , 0) #doesn't work on object with a lot of 

_, thresh = cv.threshold(blur_img, 60, 255, cv.THRESH_BINARY)   #figuring out the background. 80 is lowerest value
```

```python
plt.imshow(thresh , cmap= "gray")
```

```python
contours,_ = cv.findContours(thresh , cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE) #finding out the 
```

```python
objects = []
for cnt in contours:
    if cv.contourArea(cnt) > 1000:
      objects.append(cnt)
```

```python
object_img = img2.copy()
cv.drawContours(object_img, objects, -1, (0,255,0) , 3)
plt.imshow(object_img)
```
