```python
import cv2 as cv
import matplotlib.pyplot as plt
print(cv.__version__)
```

```python
img = cv.imread("example_03.png")
plt.imshow(img)
```

```python
gray = cv.cvtColor(img , cv.COLOR_BGR2GRAY)
plt.imshow(gray , cmap = "gray")
```

```python
blur_img = cv.GaussianBlur(gray , (5,5) , 0)
_, thresh = cv.threshold(blur_img, 100, 255, cv.THRESH_BINARY)   #figuring out the background. 80 is lowerest value
```

```python
plt.imshow(thresh , cmap = "gray")
```
